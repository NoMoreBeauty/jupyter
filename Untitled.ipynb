## 不用框架实现线性网络  
### step1：自定义数据集
### step2：构造 DataLoader
### step3：初始化参数
### step4：定义网络
### step5：定义损失函数
### step6：定义优化算法（随机梯度下降SGD）
### step7：训练

```python
#自定义数据集
def data_creater(w, b, num):
    X = torch.normal(0, 1, size=(num, len(w)))
    Y = torch.matmul(X, w) + b
    Y += torch.normal(0,0.01,size=Y.shape)
    return X,Y.reshape(-1,1) #变成列向量更好演示

true_w = torch.tensor([2, -3.4])
true_b = 3.2
X,Y = data_creater(true_w, true_b, 1000)
X,Y
```

```python
#构造 DataLoader
def DataLoader(batch_size, X, Y):
    num_examples = len(X)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i:min(i + batch_size, num_examples)])
        yield X[batch_indices], Y[batch_indices]  #返回成迭代器

batch_size = 10        
for x,y in DataLoader(batch_size, X, Y):
    print(x,'\n',y)
    break
```

```python
#初始化参数
w = torch.normal(0,0.01,size=(2,1),requires_grad=True) #后续 Xw相乘时对应维数正确 因此需要2*1不是1*2
b = torch.zeros(size=[1],requires_grad=True) # 一维的时候需要用[]
w,b
```

```python
#定义网络
def net(w, X, b):
    return torch.matmul(X,w) + b
```

```python
#定义损失函数
def loss(y_hat, y, batch_size):
    return (y_hat - y)**2 / 2 / batch_size
```

```python
#定义优化算法（随机梯度下降）
def sgd(W, lr):
    with torch.no_grad():
        for w in W:  # w和b都需要更新
            w -= lr*w.grad
            w.grad.zero_()
```

```python
#训练
epoch = 5
lr = 0.03
net_t = net
loss_t = loss

for e in range(epoch):
    for x,y in DataLoader(batch_size, X, Y):
        y_hat = net_t(w,x,b)
        l = loss_t(y_hat, y, batch_size)
        l.sum().backward() #sum对求导没有影响 sum后成为标量更方便求导（torch的规定？）
        sgd([w,b],lr)
    with torch.no_grad():
        train_l = loss(net(w, X, b), Y, batch_size)
        print(f'epoch {e + 1}, loss {train_l.mean()}')

```

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}')
```

### 交叉熵损失函数  
交叉熵损失函数一般用于概率分类，比起平方误差更关心分类结果的正确性  
$CrossEntropyLoss = -\sum_1^n$ $y_i * log \bar{y}_i$  
表示一个样本的输出概率$[0.1,0.2,0.7]$ （经过softmax）与真实概率$[0,0,1]$之间的损失  
因为简单分类中不是此类别的概率都是 0 ，因此可以简化成 $-y_ilog \bar{y}_i$  
一批量的数据的损失函数即为 $L = \frac{1}{B} \sum_1^B CrossE... $

### softmax计算
softmax计算公式：  
$\bar{y_i} = \frac{e^{o_i}}{\sum_1^n e^{o_j}}$  
第i类别的概率等于 exp(输出值) / $\sum$exp(输出值)

```python
#数据集采用 FashionMnist
import torchvision
from torch.utils import data
from torchvision import transforms
trans = transforms.ToTensor()
mnist_train = torchvision.datasets.FashionMNIST(root="../data", train=True,
                                                transform=trans,
                                                download=True)
mnist_test = torchvision.datasets.FashionMNIST(root="../data", train=False,
                                               transform=trans, download=True)

len(mnist_train), len(mnist_test)
```

```python
batch_size=32
train_iter = data.DataLoader(mnist_train, batch_size, shuffle=True)
test_iter = data.DataLoader(mnist_test, batch_size, shuffle=True)
# X, y = next(iter(data.DataLoader(mnist_train, batch_size=18)))
# X.shape
```

```python
from torch import nn
net = nn.Sequential(nn.Flatten(),nn.Linear(784,10))
net[1].weight.data.normal_(0,0.01)
net[1].bias.data.fill_(0)
```

```python
#loss = nn.CrossEntropyLoss()  #自带softmax
def softmax(y_hat):   # 32行 每行10列
    y_exp = torch.exp(y_hat)
    p = y_exp.sum(axis = 1,keepdim=True) #为了广播  32行 1列
    return y_exp / p    
```

因为标签数据是只有一个类别 例如一个batch_size的y是 [2,3,4,3,9,1...]  
因此需要一个映射操作，下述的神奇索引可以做到

```python
y = torch.tensor([0, 2])  
y_hat = torch.tensor([[0.1, 0.3, 0.6], [0.3, 0.2, 0.5]])  
y_hat[range(2), y]  
```

```python
def CrossLoss(y_hat, y):
    return -torch.log(y_hat[range(len(y_hat)), y])
#返回的是一个batch_size的损失值
```

```python
#求出一个batch中预测正确的数量
def accuracy(y_hat, y):  
    """计算预测正确的数量。"""
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat = y_hat.argmax(axis=1)
    cmp = y_hat.type(y.dtype) == y
    return float(cmp.type(y.dtype).sum())
```

```python
#sgd = torch.optim.SGD(net.parameters(), lr=0.1)
#定义优化算法（随机梯度下降）
def sgd(W, lr):
    with torch.no_grad():
        for w in W:  # w和b都需要更新
            w -= lr*w.grad
            w.grad.zero_()
```

```python
class Accumulator:  
    """在`n`个变量上累加。"""
    def __init__(self, n):
        self.data = [0.0] * n

    def add(self, *args):
        self.data = [a + float(b) for a, b in zip(self.data, args)]

    def reset(self):
        self.data = [0.0] * len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

def evaluate_accuracy(net, data_iter):  
    """计算在指定数据集上模型的精度。"""
    if isinstance(net, torch.nn.Module):
        net.eval()
    metric = Accumulator(2)
    for X, y in data_iter:
        metric.add(accuracy(net(X), y), y.numel())
    return metric[0] / metric[1]

```

```python
test_acc = evaluate_accuracy(net, test_iter)
print(test_acc)
```

```python
#训练
epoch=10
for e in range(epoch):
    test_acc = evaluate_accuracy(net, test_iter)
    for x,y in train_iter:
        l = CrossLoss(softmax(net(x)),y)
        l = l.sum()/batch_size
        l.backward()
        sgd([net[1].weight,net[1].bias],0.1)
    print(test_acc)
    
```

##### 1 参数数量
##### 2 参数取值范围

```python
import torch
import numpy as np
```

通过拟合多项式演示过拟合与欠拟合

### step1：创建数据集
### step2：评估方法
### step3：训练

$y = 5 + 1.2x - 3.4\frac{x^2}{2!} + 5.6\frac{x^3}{3!} + e(噪音)$

```python
# step1
import math
max_degree = 20
n_train, n_test = 100, 100
true_w = np.zeros(max_degree)
true_w[0:4] = np.array([5, 1.2, -3.4, 5.6])
#前四维正常 后16维为0

features = np.random.normal(size=(n_train + n_test, 1))  #生成200个数据
np.random.shuffle(features)
poly_features = np.power(features, np.arange(max_degree).reshape(1, -1)) #数据通过广播机制先形成  n次方的向量
for i in range(max_degree):
    poly_features[:, i] /= math.gamma(i + 1)  #除阶乘
labels = np.dot(poly_features, true_w)  # 多项式求和等于求点积
labels += np.random.normal(scale=0.1, size=labels.shape) #噪音

#转成 tensor
true_w, features, poly_features, labels = [
    torch.tensor(x, dtype=torch.float32)
    for x in [true_w, features, poly_features, labels]]
```

```python
#step2
class Accumulator:  
    """在`n`个变量上累加。"""
    def __init__(self, n):
        self.data = [0.0] * n

    def add(self, *args):
        self.data = [a + float(b) for a, b in zip(self.data, args)]

    def reset(self):
        self.data = [0.0] * len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

def evaluate_loss(net, data_iter, loss):  
    """评估给定数据集上模型的损失。"""
    metric = Accumulator(2)
    for X, y in data_iter:
        out = net(X)
        y = y.reshape(out.shape)
        l = loss(out, y)
        metric.add(l.sum(), l.numel())
    return metric[0] / metric[1]
```

```python
#step3
from torch import nn
import torchvision
from torch.utils import data
from torchvision import transforms
#使用API DATALOADER
def load_arr(data_arrays, batch_size, is_train=True):
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size,shuffle = is_train)

def train(train_features, test_features, train_labels, test_labels,
          num_epochs=400):
    loss = nn.MSELoss()
    input_shape = train_features.shape[-1] 
    net = nn.Sequential(nn.Linear(input_shape, 1, bias=False))
    batch_size = min(10, train_labels.shape[0])
    train_iter = load_arr((train_features, train_labels.reshape(-1, 1)),batch_size,is_train=True)
    test_iter = load_arr((test_features, test_labels.reshape(-1, 1)),batch_size, is_train=False)
    sgd = torch.optim.SGD(net.parameters(), lr=0.01,weight_decay=0.01) #后面实验为了看看L2正则加的weight_decay,效果居然还很不错
    l_l = []
    for epoch in range(num_epochs):
        for x,y in train_iter:
            l = loss(net(x),y)
            sgd.zero_grad()
            l.backward()
            sgd.step()
        l_l.append(evaluate_loss(net,test_iter,loss))
    return l_l,net[0].weight.data.numpy()
```

## 解决过拟合的手段
### ① L2正则(权重衰退)
### ② Dropout

#### L2正则
过拟合的原因之一是模型参数复杂，参数复杂的一个方面是参数范围问题，因此改善过拟合的一个方法是限制参数范围  
因此理论上可以添加硬性规定  
$loss = min\{l(w,b)\}  其中 ||w|| < \theta$  
这样的限制实现比较困难，数学上，这一条限制等价于  
$loss = min\{l(w,b) + \frac{\lambda}{2}||w||^2 \}$  
可通过拉格朗日乘子证明  
这里的$\lambda$反比对应$\theta$ （$\lambda$越大，参数范围越小） 
直观上为了最小化loss，就不能让权重w的范数增大，因此限制了参数范围
##### 又被称为权重衰退
因为求导后可化简为 
$w_{t+1} = (1 - \eta\lambda)w_t - \eta \frac{\partial loss}{\partial w}$  
一般情况下，$\eta\lambda < 1 $
因此每次更新权重时都相当于对其进行了衰减，也被称为权重衰减（可能造成梯度消失？）

```python
import torch
```

简单实现

```python
from torch import nn
net = nn.Sequential(nn.Linear(10, 1, bias=False))
sgd = torch.optim.SGD(net.parameters(), lr=0.01, weight_decay=0.01)
#这里的weight_decay就是优化器集成的L2正则，只有L2正则，0.01是lambda超参数
#不使用优化器集成的L2正则就要自己写函数加在求导或者求loss前
```

#### Dropout
Dropout为什么可以改善过拟合数学上可以把Dropout看成一种正则项  
Dropout一般用在全连接层的输出上（Relu过后进Dropout）  
Dropout的要求是 x的期望 E(x)不改变，因此满足公式  
$x = \left\{
\begin{matrix}
0概率=p \\
\frac{x}{1-p} 概率为1-p 
\end{matrix}
\right.$  
所有的正则式只在训练时使用

简单实现

```python
from torch import nn
p=0.5
net = nn.Sequential(nn.Linear(10, 1),nn.ReLU(),nn.Dropout(p))
```
